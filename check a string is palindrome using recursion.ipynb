```python
def checking(li, left, right):
    if left >= right:
        return True  # all checks passed
    
    if li[left] != li[right]:
        return False  # mismatch â†’ not palindrome
    
    return checking(li, left + 1, right - 1)
```

```python
def check1(li, left, right):
    while left < right:
        if li[left] != li[right]:
            return False
        left += 1
        right -= 1
    return True
```
