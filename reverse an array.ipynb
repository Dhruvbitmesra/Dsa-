```python
##reverse an array using recurssion
```

```python
def rev(arr, left, right):##normal reversing
    if left > right:
        return arr
    
    while left < right:
        arr[left], arr[right] = arr[right], arr[left]  # swap
        left += 1
        right -= 1
    
    return arr
```

```python
num=[1,2,3,4,5]
revv(num,0,4)
```

```python
def revv(arr, left, right):##recursion reverse array
    if left >= right:
        return arr
    
    arr[left], arr[right] = arr[right], arr[left]
    return rev(arr, left + 1, right - 1)
```
