```python
##hashing
'''prestoring values into some datastructure like list dict tuples and then fetching it 

lets take a question to understand'''

n = [5,3,2,2,1,5,5,7,5,10]
m = [10,11,1,9,5,6,7,2]

for num in m:
    count = 0
    for x in n:
        if x == num:
            count += 1
    print(num, "→", count)
```

```python
##hash list approach
##useful only when 1<=n[i]<=10


n = [5,3,2,2,1,5,5,7,5,10]
m = [10,11,1,9,5,6,7,2]

hash_list = [0] * 11   # index 0 unused

for num in n:
    hash_list[num] += 1

for num in m:
    if num < 1 or num > 10:
        print(0)
    else:
        print(hash_list[num])
```

```python
##frequency dict approach 


num = [5,3,2,2,1,5,5,7,5,10]
mum = [10,11,1,9,5,6,7,2]

hash_map = {}
result = []

# Build frequency map
for x in num:
    hash_map[x] = hash_map.get(x, 0) + 1

print("Frequency map:", hash_map)

# Query map
for j in mum:
    result.append(hash_map.get(j, 0))

print("Result:", result)
```

```python
##character hashing

s="azyxyyzaaaa"
q=["d","a","y","x"]


hash_list=[0]*26


for ch in s:
    ascii_val=ord(ch)
    index=ascii_val-97
    hash_list[index]+=1
    
for ch in q:
    ascii_val=ord(ch)
    index=ascii_val-97
    print(hash_list[index])

```
