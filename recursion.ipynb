```python
def func(count=0):
    if count == 4:
        return
    print("Dhruv")
    func(count + 1)

func()
```

```python
def func1(count=0):
    if count == 4:
        return
    count += 1
    func1(count)
    print("dhruv")
```

```python
def func2(x,n):
    if n==0:
        return
    print(x)
    func2(x,n-1)
```

```python
func2(15,4)
```

```python
def printing(n):
    if n==0: ##base case
        return
    print(n)
    printing(n-1)
```

```python
printing(10)
```

```python
def haidrec(i,n):
    if i>n:
        return
    haidrec(i+1,n)
    print(i)
 
```

```python
haidrec(1,5)
```

```python
def tailrec(i,n):
    if i>n:
        return
    print(i)
    tailrec(i+1,n)
 
```

```python
tailrec(1,5)
```

```python
##parameterized and functional recursion
```

```python
def sumfn(sum,i,n):##parameterized recursion
    if i>n:
        print(sum)
        return
    sumfn(sum+i,i+1,n)
```

```python
sumfn(0,1,10)
```

```python
##functional recursion (we dont print we return something)
def funcrec(n):
    if n==1:##base condition
        return 1 
    return n+funcrec(n-1)##recurrence relation
```

```python
funcrec(10)
```

```python
##factorial 

def factorial(n):
    if n==0 or n==1:
        return 1
    return n*factorial(n-1)
```

```python
factorial(5)
```
