```python
n = 1234
num = n
result = 0

while num > 0:
    ld = num % 10
    result = (result * 10) + ld
    num = num // 10

print(n == result)
```

```python
##armstrong number
n = 153
num = n
result = 0
nod=len(str(num))
while num > 0:
    last_digit = num % 10
    result = result + (last_digit ** nod)
    num = num // 10

print(n == result)
```

```python
from math import sqrt


n=36
result=[]

for i in range(1,int(sqrt(n))+1):
    if n%i==0:
        result.append(i)
        if n//i !=i:
            result.append(n//i)

result.sort()

print(result)
    
```

```python
nums=[5,6,7,7,22,11,3,4,56,7]

freq_map=dict()
for i in range (0,len(nums)):
    if nums[i] in freq_map:
        freq_map[nums[i]]+=1
    else:
        freq_map[nums[i]]=1


print(freq_map)
```

```python
hash_map={}
nums=[5,6,7,7,22,11,3,4,56,7]

n=len(nums)

for i in range(0,n):
    hash_map[nums[i]]=hash_map.get(nums[i],0)+1

print(hash_map)
```
