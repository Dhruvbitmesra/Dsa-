```python
name = 'nitin'
flag = True

for i in range(len(name)):
    if name[i] != name[len(name) - 1 - i]:
        flag = False
        break

if flag:
    print("yes, palindrome")
else:
    print("no")
```

```python
n = 1234
num = n
result = 0

while num > 0:
    ld = num % 10
    result = (result * 10) + ld
    num = num // 10

print(n == result)
```

```python
##armstrong number
n = 153
num = n
result = 0
nod=len(str(num))
while num > 0:
    last_digit = num % 10
    result = result + (last_digit ** nod)
    num = num // 10

print(n == result)
```
